```python
import matplotlib.pyplot as plt
import numpy as np

def linear_regression(X, Y, L, number_of_iterations, color):
    theta0 = 0
    theta1 = 0

    m = float(len(X))

    for i in range(number_of_iterations):
        Y_pred = theta1 * X + theta0
        D_theta0 = (1/m) * sum(Y_pred - Y)
        D_theta1 = (1/m) * sum((Y_pred - Y) * X)
        theta0 = theta0 - L * D_theta0
        theta1 = theta1 - L * D_theta1

    Y_pred = theta0 + theta1 * X

    plt.scatter(X, Y)
    plt.plot(X, Y_pred, color=color, label=f'Line: L={L}')

linear_regression(X, Y, 0.001, 5000, color='blue')
linear_regression(X, Y, 0.003, 5000, color='black')
linear_regression(X, Y, 0.009, 5000, color='red')

plt.legend()
plt.show()
```
